## Part D: Algorithm 1

```python
import numpy as np
import scipy
import sklearn.datasets
import matplotlib.pyplot as plt
%matplotlib inline

X_train, y_train = sklearn.datasets.load_svmlight_file("train.txt")
X = X_train.todense()
```

```python
def algorithm1(X,y,Imax,w_init,α):
    w = w_init
    i = 0
    while i<Imax:
        w_temp = 0
        for j in range(X.shape[0]):
            if y[j]*w.dot(np.transpose(X[j])) <0:
                w_temp = w_temp - y[j]*X[j]
        w = w - α*w_temp
        i = i +1
    return w

def h(w):
    #returns slope of separating hyperplane associated with normal vector w
    return w[0,0]/(-w[0,1])
```

```python
Imax = 2000
w0 = np.array([1.0,1.0])

w_6 = algorithm1(X,y_train,Imax,w0,0.006)
w_7 = algorithm1(X,y_train,Imax,w0,0.007)
w_8 = algorithm1(X,y_train,Imax,w0,0.008)
w_9 = algorithm1(X,y_train,Imax,w0,0.009)
w_10 = algorithm1(X,y_train,Imax,w0,0.010)
w_11 = algorithm1(X,y_train,Imax,w0,0.011)
w_12 = algorithm1(X,y_train,Imax,w0,0.012)
w_13 = algorithm1(X,y_train,Imax,w0,0.013)
w_14 = algorithm1(X,y_train,Imax,w0,0.014)
w_15 = algorithm1(X,y_train,Imax,w0,0.015)



```

```python
grid = range(0,11,5)
h6 = h(w_6)*grid
h7 = h(w_7)*grid
h8 = h(w_8)*grid
h9 = h(w_9)*grid
h10 = h(w_10)*grid
h11 = h(w_11)*grid
h12 = h(w_12)*grid
h13 = h(w_13)*grid
h14 = h(w_14)*grid
h15 = h(w_15)*grid

```

```python
x =np.asarray(X[:,0])
y = np.asarray(X[:,1])
plt.scatter(x,y)
plt.plot( grid, h10, linewidth=2, linestyle=':', label = "h10")
plt.plot( grid, h15, linewidth=2, linestyle='--', label = "h15" )
plt.plot( grid, h6, linewidth=2, linestyle='-', label = "h6")         
plt.title("Algorithm 1")
plt.xlabel('Attribute 1')
plt.ylabel('Attribute 2')
plt.legend()
plt.savefig('algorithm1.png')
plt.show()

```

## Part D: Algorithm 2

```python
def algorithm2(X,y,Imax,w_init,α,C):
    w = w_init
    i = 0
    while i<Imax:
        w_temp = w
        for j in range(X.shape[0]):
            if y[j]*w.dot(np.transpose(X[j]))>1:
                w_temp = w_temp - C*y[j]*X[j]
        w = w - α*w_temp
        i = i +1
    return w
```

```python
C = 0.01

w2_6 = algorithm2(X,y_train,Imax,w0,0.006,C)
w2_7 = algorithm2(X,y_train,Imax,w0,0.007,C)
w2_8 = algorithm2(X,y_train,Imax,w0,0.008,C)
w2_9 = algorithm2(X,y_train,Imax,w0,0.009,C)
w2_10 = algorithm2(X,y_train,Imax,w0,0.010,C)
w2_11 = algorithm2(X,y_train,Imax,w0,0.011,C)
w2_12 = algorithm2(X,y_train,Imax,w0,0.012,C)
w2_13 = algorithm2(X,y_train,Imax,w0,0.013,C)
w2_14 = algorithm2(X,y_train,Imax,w0,0.014,C)
w2_15 = algorithm2(X,y_train,Imax,w0,0.015,C)

h6 = h(w2_6)*grid
h10 = h(w2_10)*grid
h15 = h(w2_15)*grid

plt.scatter(x,y)
plt.plot( grid, h10, linewidth=2, linestyle=':', label = "h10")
plt.plot( grid, h15, linewidth=2, linestyle='--', label = "h15" )
plt.plot( grid, h6, linewidth=2, linestyle='-', label = "h6")         
plt.title("Algorithm 2")
plt.xlabel('Attribute 1')
plt.ylabel('Attribute 2')
plt.legend()
plt.savefig('algorithm2.png')
plt.show()
```

## Part E

```python
def algorithm3(X,y,Imax,w_init,α):
    w = w_init
    i = 0
    update = True
    num_passes = 0
    while i<Imax and update==True:
        update = False
        for j in range(X.shape[0]):
            if y[j]*w.dot(np.transpose(X[j]))<=0:
                w = w + α*y[j]*X[j]
                i = i +1
                update = True
        num_passes = num_passes + 1
    return w,i,num_passes
```

```python
Xperc_train, yperc_train = sklearn.datasets.load_svmlight_file("train_perc.txt")
Xperc = Xperc_train.todense()

Xpercsort_train, ypercsort_train = sklearn.datasets.load_svmlight_file("train_perc_sorted.txt")
Xpercsort = Xpercsort_train.todense()

```

```python
Imax = 2000
α = 0.01
w0 = np.array([0.0,0.0])

w_unsort,i,num_passes = algorithm3(Xperc,yperc_train,Imax,w0,α)

print('The number of updates was ',str(i),' and the number of passes was ',str(num_passes))
```

```python
w_sort,i,num_passes = algorithm3(Xpercsort,ypercsort_train,Imax,w0,α)

print('The number of updates was ',str(i),' and the number of passes was ',str(num_passes))
```
